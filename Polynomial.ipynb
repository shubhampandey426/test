```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"C:\Users\win10\Downloads\Polynomial_Regression_and_Data_Transformation\polynom.csv")
data.head()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=data,x='number_courses',y='Marks')
plt.title('Number Of Course VS Marks')
plt.show()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(data=data,x='time_study',y='Marks')
plt.title('Time Study VS Marks')
plt.show()
```

```python
# Spliting the Data into X and y
y=data['Marks']
X=data.drop('Marks',axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,random_state=42)
```

```python
# Building Model
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score 
```

```python
r2_score(y_test,y_pred)
```

```python
y_full_pred=reg.predict(X)
```
